### Which three years of the 1930s saw the most films released?

```python

df = pd.DataFrame(file)

df_filtered = df[df["year"].between(1930, 1939)]

year_counts = df_filtered["year"].value_counts()

year_counts_sorted = year_counts.sort_values(ascending=False)

top_three_years = year_counts_sorted.head(3)
print(top_three_years)
```

### Plot the number of films that have been released each decade over the history of cinema.

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(file)

df["decade"] = (df["year"] // 10) * 10
decade_counts = df["decade"].value_counts()

plt.figure(figsize=(10,4))
decade_counts.plot(kind="bar", color=plt.cm.viridis.colors)
plt.xlabel("Decade")
plt.ylabel("Number of movies")
plt.title("Number of Films Released Each Decade")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### What are the 11 most common character names in movie history?

### Which actors or actresses appeared in the most movies in the year 1945?

```python
df_filtered = df[df['year'] == 1945]
actor_counts = df_filtered.groupby('name')['title'].count().reset_index(name='count')
sorted_actor_counts = actor_counts.sort_values(by='count', ascending=False)
most_movies_actors = sorted_actor_counts.head(1)

print("The actors or actresses 5:",)
for index, row in most_movies_actors.iterrows():
    print(f"{row['name']}: {row['count']} movies")
```

### What are the 10 most frequent roles that start with the phrase "Patron in"?

### How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
leading_roles_1950s = df[(df['n'] == 1) & (df['year'] >= 1950) & (df['year'] < 1960)]

actor_count = leading_roles_1950s[leading_roles_1950s['type'] == "actor"].shape[0]
actress_count = leading_roles_1950s[leading_roles_1950s['type'] == "actress"].shape[0]
print("Number of leading roles for actors in the 1950s:", actor_count)
print("Number of leading roles for actresses in the 1950s:", actress_count)
```
